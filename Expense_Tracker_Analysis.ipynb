```python
## Data Loading
```

```python
import pandas as pd

data = {
    'Date': [
        '2024-01-01','2024-01-02','2024-01-03','2024-01-04','2024-01-05',
        '2024-01-06','2024-01-07','2024-01-08','2024-01-09','2024-01-10',
        '2024-01-11','2024-01-12','2024-01-13','2024-01-14','2024-01-15'
    ],
    'Category': [
        'Food','Transport','Shopping','Food','Bills',
        'Entertainment','Food','Transport','Food','Shopping',
        'Bills','Entertainment','Food','Transport','Shopping'
    ],
    'Amount': [
        250, 100, 1200, 300, 2000,
        500, 350, 150, 400, 800,
        1800, 600, 300, 120, 950
    ]
}

df = pd.DataFrame(data)
df
```

```python
df.info()
df.describe()
```

```python
## Data Cleaning
```

```python
## Exploratory Data Analysis
```

```python
### Total Spending by Category
```

```python
df.groupby('Category')['Amount'].sum().plot(kind='bar')
```

```python
df.sort_values(by='Amount', ascending=False).head()
```

```python
### Category Distribution
```

```python
df.groupby('Category')['Amount'].sum().plot(kind='pie', autopct='%1.1f%%')
```

```python
df.groupby('Category')['Amount'].sum().sort_values(ascending=False)
```

```python
df.groupby('Category')['Amount'].mean()
```

```python
df.groupby('Category')['Amount'].sum().idxmax()
```

```python
monthly_total = df['Amount'].sum()
monthly_total
```

```python
### Daily Spending Trend
```

```python
df.set_index('Date')['Amount'].plot()
```

```python
## Insights

# A significant portion of expenses is concentrated in categories like Bills and Shopping.

# Food and Transport expenses occur frequently, indicating regular daily spending.

# Bills contribute the highest total expense, showing major fixed costs.

# Some categories have high average spending, even if they occur less frequently.

# Spending patterns suggest opportunities to reduce non-essential expenses like entertainment and shopping.
```
